```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
data = pd.read_csv('/content/Wine_clust (2).csv')
# Load the Iris dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
def plot_clusters(X, labels, title):
    plt.figure(figsize=(8, 6))
    plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50, alpha=0.5)
    plt.title(title)
    plt.xlabel('PCA Component 1')
    plt.ylabel('PCA Component 2')
    plt.colorbar()
    plt.show()
```

```python
def kmeans_clustering(X, max_clusters=10):
    silhouette_scores = []
    for n_clusters in range(2, max_clusters + 1):
        kmeans = KMeans(n_clusters=n_clusters, random_state=42)
        kmeans.fit(X)
        labels = kmeans.labels_
        silhouette_avg = silhouette_score(X, labels)
        silhouette_scores.append(silhouette_avg)

    best_n_clusters = np.argmax(silhouette_scores) + 2  # Add 2 to start from 2 clusters
    kmeans = KMeans(n_clusters=best_n_clusters, random_state=42)
    kmeans.fit(X)
    labels = kmeans.labels_
    plot_clusters(X, labels, f'K-means Clustering (Optimal number of clusters: {best_n_clusters})')
```

```python
def agglomerative_clustering(X, max_clusters=10):
    silhouette_scores = []
    for n_clusters in range(2, max_clusters + 1):
        agg_clustering = AgglomerativeClustering(n_clusters=n_clusters)
        agg_clustering.fit(X)
        labels = agg_clustering.labels_
        silhouette_avg = silhouette_score(X, labels)
        silhouette_scores.append(silhouette_avg)

    best_n_clusters = np.argmax(silhouette_scores) + 2  # Add 2 to start from 2 clusters
    agg_clustering = AgglomerativeClustering(n_clusters=best_n_clusters)
    labels = agg_clustering.fit_predict(X)
    plot_clusters(X, labels, f'Agglomerative Clustering (Optimal number of clusters: {best_n_clusters})')
```

```python
def dbscan_clustering(X, eps=0.5, min_samples=5):
    dbscan = DBSCAN(eps=eps, min_samples=min_samples)
    labels = dbscan.fit_predict(X)
    plot_clusters(X, labels, 'DBSCAN Clustering')

# Optimum number of clusters for K-means
kmeans_clustering(X_scaled)

# Optimum number of clusters for Agglomerative clustering
agglomerative_clustering(X_scaled)

# DBSCAN clustering
dbscan_clustering(X_scaled)
```
